```python
# reading data
import pandas as pd

Billionaires = pd.read_csv('/content/drive/MyDrive/datasets/Forbes Billionaires Data v2.csv')
Billionaires.head()
```

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
from plotnine import *
```

```python
# summarizing a column of data, eg Companies
Companies = Billionaires['Companies'].value_counts().reset_index()
Companies = Companies.head(10)
```

```python
import matplotlib.pyplot as plt
import pandas as pd



# Plot
plt.figure(figsize=(10,6))
plt.bar(Companies["Companies"], Companies["count"])

# Aesthetics
plt.title("Company Participation in 2016")
plt.xlabel("Participation Status")
plt.ylabel("Number of Companies")
plt.xticks(rotation = 90)

# Display the chart
plt.show()
```

```python
# variables that are continuous are the ones plotted in a histogram
import matplotlib.pyplot as plt

# Count number of billionaires per country
country_counts = Billionaires["Country"].value_counts().reset_index()
country_counts = country_counts.sort_values(by = "count", ascending = True).tail(10)
# Plot histogram-like bar chart of the counts
plt.figure(figsize=(10,6))
plt.barh(country_counts['Country'], country_counts['count'], color='skyblue')

# Aesthetics
plt.title("Distribution of Billionaires Across Countries")
plt.xlabel("Number of Billionaires per Country")
plt.ylabel("Frequency (Number of Countries)")
# plt.tight_layout()
plt.show()
```

```python
# importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
from plotnine import *
```

```python
net_average = Billionaires.groupby("Companies")['Net Worth'].mean().to_frame().reset_index().round(1)
net_average = net_average.sort_values(by = 'Net Worth', ascending = False)
net_average
```

```python
# plotting using seaborn
plt.figure(figsize=(10, 7))
sns.barplot(x = 'Companies', y = 'Net Worth', data = net_average.head(10), palette = "viridis")

# aesthetics
plt.xticks(rotation = 90)
plt.show()
```
